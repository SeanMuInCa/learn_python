Which country has won the most men's gold medals in singles badminton over the years? Sort the results alphabetically by the player's names.(2 points)

```python
dataWonMenBadmintonSingle = data[(data.Medal == 'Gold') & (data.Gender == 'Men') & (data.Sport == 'Badminton') & (data.Event == 'singles')]
dataWonMenBadmintonSingle['NOC'].value_counts().head(1)
dataWonMenBadmintonSingle.sort_values(by='Athlete')
```

Which three countries have won the most medals in recent years (from 1984 to 2008)? (2 points)

```python
data[(data.Edition >= 1984) & (data.Edition <= 2008)]['NOC'].value_counts().head(3)
```

Display the male gold medal winners for the 100m Track event over the years. List the results starting with the most recent. Show the Olympic City, Edition, Athlete and the country they represent. (2 points)

```python
data[(data.Medal == 'Gold') & (data.Gender == 'Men') & (data.Event == '100m')].sort_values(by='Edition', ascending=False)[['City','Edition','Athlete','NOC']]
```

Which countries did not win a medal in the 2008 Olympics? How many countries were there? (2 points)

```python
metalCountriesIn2008 = data[data.Edition == 2008]['NOC']
allCountries = data.drop_duplicates(subset=['NOC'])['NOC']
resMissing = allCountries[~allCountries.isin(metalCountriesIn2008)]
resMissing
resMissing.count()
```

Using groupby(), plot the total number of medals awarded at each of the Olympic games throughout history. (2 points)

```python
data.groupby('Event').count()['Medal']
```

Create a list showing the total number of medals won for each country over the history of the Olympics. For each country, include the year of the first and most recent Olympic medal wins. (3 points)

```python
total_medals = data.groupby('NOC')['Medal'].count()
first_medal_year = data.groupby('NOC')['Edition'].min()
recent_medal_year = data.groupby('NOC')['Edition'].max()
medal_summary = pd.DataFrame({
    'Total Medals': total_medals,
    'First Medal Year': first_medal_year,
    'Recent Medal Year': recent_medal_year
})
medal_summary
```

Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 using:
- matplotlib
- Seaborn      (3 points)

```python
medals_count = data[(data.NOC == 'CHN') & (data.Edition == 2008)].groupby('Gender')['Medal'].count()
import matplotlib.pyplot as plt


medals_count.plot(kind='bar', stacked=True)

plt.title('Number of Medals Achieved by Chinese team in Beijing 2008')
plt.xlabel('Gender of Chinese team')
plt.ylabel('Number of Medals')

plt.show()
```
