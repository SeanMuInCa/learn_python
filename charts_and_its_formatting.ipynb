```python
# Importing Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt

# Configuring Matplotlib for better visuals
plt.style.use("seaborn-v0_8")

```

#various charts

Line Chart: A line chart is used to display information as a series of data points connected by a straight line. It is often used for trend analysis.

```python
# Data for the line chart
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotting the line chart
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sine Wave', color='blue')
plt.title('Line Chart Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

Bar Chart: A bar chart displays categorical data using rectangular bars, where the height (or length) represents the value.

```python
# Data for the bar chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# Plotting the bar chart
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

Horizontal Bar Chart: Similar to a bar chart but the bars are horizontal.

```python
# Plotting the horizontal bar chart
plt.figure(figsize=(8, 5))
plt.barh(categories, values, color='lightgreen')
plt.title('Horizontal Bar Chart Example')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

Scatter Plot: A scatter plot uses dots to represent values for two variables. It’s useful for identifying relationships or patterns.

```python
# Data for the scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 500 * np.random.rand(50)

# Plotting the scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Color Intensity')
plt.show()
```

Histogram: A histogram is used to represent the distribution of data by dividing it into bins.

```python
# Data for the histogram
data = np.random.randn(1000)

# Plotting the histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

 Pie Chart: A pie chart is used to show proportions of a whole.

```python
# Data for the pie chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# Plotting the pie chart
plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Box Plot: A box plot is used to show the distribution of a dataset and highlights outliers.

```python
# Data for the box plot
data = [np.random.randn(100), np.random.randn(100) + 1]

# Plotting the box plot
plt.figure(figsize=(8, 5))
plt.boxplot(data, vert=True, patch_artist=True, labels=['Group 1', 'Group 2'])
plt.title('Box Plot Example')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()
```

Heatmap (Using imshow): A heatmap represents data in a matrix format using color gradients.

```python
# Data for the heatmap
data = np.random.rand(10, 10)

# Plotting the heatmap
plt.figure(figsize=(8, 5))
plt.imshow(data, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Value')
plt.title('Heatmap Example')
plt.show()
```

# Chart Formatting

```python
# Example: Formatting a Line Chart

# Data for the chart
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and axis
plt.figure(figsize=(10, 6))  # Setting figure size

# Plot multiple lines with custom styles
plt.plot(x, y1, label='Sine Wave', color='blue', linewidth=2, linestyle='--', marker='o', markersize=6)
plt.plot(x, y2, label='Cosine Wave', color='red', linewidth=2, linestyle='-', marker='s', markersize=6)

# Adding Title with Formatting
plt.title('Formatted Line Chart', fontsize=16, fontweight='bold', color='darkgreen', pad=20)

# Adding Axis Labels with Formatting
plt.xlabel('X-axis Label', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Y-axis Label', fontsize=14, fontweight='bold', color='darkblue')

# Customizing the Legend
plt.legend(title='Legend', title_fontsize=12, fontsize=10, loc='upper right', frameon=True, shadow=True)

# Adding Gridlines
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Customizing Ticks
plt.xticks(fontsize=12, rotation=45, color='purple')
plt.yticks(fontsize=12, color='purple')

# Adding Annotations
plt.annotate('Peak', xy=(1.57, 1), xytext=(2.5, 1.2),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=12, color='black')

# Adding a Background Color to the Plot
plt.gca().set_facecolor('lightyellow')

# Saving the Chart as an Image
plt.savefig('formatted_line_chart.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
