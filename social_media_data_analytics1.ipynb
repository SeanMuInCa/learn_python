# **Filling Missing Default Language as English**

# **Converting Duration into Seconds**

```python
youtubeData['duration'] = youtubeData['duration'].str.replace('PT', '')
import re
def convertToSeconds(time_str):
    match = re.match(r'(?:(\d+)M)?(\d+)S', time_str)

    if match:
        minutes = int(match.group(1)) if match.group(1) else 0
        seconds = int(match.group(2))
        return minutes * 60 + seconds
    return None
youtubeData['duration'] = youtubeData['duration'].apply(convertToSeconds)
youtubeData.head()
```

# **Imporing TextBlob for Sentiment Analysis**

```python
from textblob import TextBlob
```

# Defining Functions

```python
def getSubjectivity (text):
    return TextBlob(text).sentiment.subjectivity

# Create a function to get the polarity
def getPolarity (text):
    return TextBlob(text). sentiment.polarity
```

# **Preprocessing Comments**

```python
youtubeData['comments'] = youtubeData['comments'].fillna('')
import ast

def convert_string_to_list(string):
    try:
        return ast.literal_eval(string)
    except:
        return []

youtubeData['comments'] = youtubeData['comments'].apply(convert_string_to_list)
```

# **Calculating Sentiment Scores**

```python
youtubeData['subjectivity'] = pd.Series(dtype=object) # will be a warning if assign this to 0 directly
youtubeData['polarity'] = pd.Series(dtype=object)

for index, row in youtubeData.iterrows():
    subjectivity_scores = []
    polarity_scores = []
    comments = row['comments']
    if comments:  # check empty
        for comment in comments:
            if comment:  # check empty
                subjectivity_scores.append(getSubjectivity(comment))
                polarity_scores.append(getPolarity(comment))
    # assign the lists of scores back
    youtubeData.at[index, 'subjectivity'] = subjectivity_scores
    youtubeData.at[index, 'polarity'] = polarity_scores


youtubeData.iloc[20:30]
```

# **Doing Polarity Analysis**

```python
# polarity analysis
youtubeData['polarityAnalysis'] = ''

for index, row in youtubeData.iterrows():
    polarity = row['polarity']
    if polarity:  # check empty
        scorePos = 0
        scoreNeg = 0
        for p in polarity:
            if p > 0:
                scorePos += 1
            elif p < 0:
                scoreNeg += 1

        # 3 of 5 to decide
        if scorePos >= 3:
            sentiment = 'positive'
        elif scoreNeg >= 3:
            sentiment = 'negative'
        else:
            sentiment = 'neutral'

        youtubeData.at[index, 'polarityAnalysis'] = sentiment
```

# **Doing Subjectivity Analysis**

```python
# subjectivity analysis
youtubeData['subjectivityAnalysis'] = ''

for index, row in youtubeData.iterrows():
    subjectivity = row['subjectivity']
    if subjectivity:  # check empty
        highSubjectivity = 0
        lowSubjectivity = 0
        midSubjectivity = 0
        for s in subjectivity:
            if s > 0.75:
                highSubjectivity += 1
            elif s <= 0.25:
                lowSubjectivity += 1
            else:
                midSubjectivity += 1
        # 3 of 5 to decide
        if highSubjectivity >= 3:
            sentiment = 'high subjectivity'
        elif lowSubjectivity >= 3:
            sentiment = 'low subjectivity'
        elif midSubjectivity >=3:
            sentiment = 'midsubjectivity'

        youtubeData.at[index, 'subjectivityAnalysis'] = sentiment
youtubeData.iloc[10:20]
```

# **Calculating Average Sentiment Scores**

```python

youtubeData['averageSubjectivity'] = pd.Series(dtype=object)
youtubeData['averagePolarity'] = pd.Series(dtype=object)

for index, row in youtubeData.iterrows():
    score1 = 0
    score2 = 0
    subjectivity = row['subjectivity']
    polarity = row['polarity']
    if subjectivity:
        for s in subjectivity:
            score1 += s
        score1 = score1 / len(subjectivity)
    if polarity:
        for p in polarity:
            score2 += p
        score2 = score2 / len(polarity)
    youtubeData.at[index, 'averageSubjectivity'] = score1
    youtubeData.at[index, 'averagePolarity'] = score2

# Display the first few rows to check the result
youtubeData.head()
```
