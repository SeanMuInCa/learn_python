```python
import pandas as pd
```

```python
covid19Confirm_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
covid19Death_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
covid19Confirm = pd.read_csv(covid19Confirm_url)
covid19Death = pd.read_csv(covid19Death_url)
```

```python
date_columns = covid19Confirm.columns[4:] # change the date format
covid19Confirm.columns = list(covid19Confirm.columns[:4]) + list(pd.to_datetime(date_columns, format='%m/%d/%y').strftime('%Y-%m-%d'))
```

```python
date_columns = covid19Death.columns[4:]# change the date format
covid19Death.columns = list(covid19Death.columns[:4]) + list(pd.to_datetime(date_columns, format='%m/%d/%y').strftime('%Y-%m-%d'))
```

```python
covid19Confirm['Province/State'] = covid19Confirm.apply(
    lambda row: row['Country/Region'] if pd.isnull(row['Province/State']) else row['Province/State'], axis=1)
```

```python
covid19Death['Province/State'] = covid19Death.apply(
    lambda row: row['Country/Region'] if pd.isnull(row['Province/State']) else row['Province/State'], axis=1)
```

```python
covid19Confirm = covid19Confirm.dropna()
```

```python
covid19Death = covid19Death.dropna()
```

```python
# define a new df to collect daily data
columns = ['Date', 'GlobalConfirmed', 'GlobalDeath']
globalCovidData = pd.DataFrame(columns=columns)

globalCovidData['Date'] = covid19Confirm.columns[4:]

sumConfirmed = 0 # daily confirm
sumDeath = 0 # daily death


for i in range(len(globalCovidData)):

    current_date = globalCovidData['Date'].iloc[i]


    sumConfirmed = covid19Confirm[current_date].sum()
    sumDeath = covid19Death[current_date].sum()


    globalCovidData.loc[i, 'GlobalConfirmed'] = sumConfirmed
    globalCovidData.loc[i, 'GlobalDeath'] = sumDeath
globalCovidData
```

```python
import pandas as pd
import yfinance as yf

# Define the stocks and their symbols
stocks = {
    'S&P500 ETF':'SPY',              #  overall american market
    'TSX Composite ETF': 'XIU.TO',        # overall canadian market
    'Air Canada': 'AC.TO',            # Air Canada, Toronto Stock Exchange - Travel sector
    'Choice Properties REIT': 'CHP-UN.TO',    # Choice Properties REIT, Toronto Stock Exchange - The Real Estate sector
    'SPDR Gold Trust ETF': 'GLD'        # SPDR Gold Trust ETF - Precious metals
}

# Define the date range (from start_date to the date before end_date)
start_date = '2020-01-22'
end_date = '2023-03-10'

# Create an empty DataFrame to store the data
stock_data = pd.DataFrame()

# Loop through the stocks
for stock_name, symbol in stocks.items():
    # Fetch historical stock data using yfinance
    data = yf.download(symbol, start=start_date, end=end_date)

    # Add 'Stock' and 'Symbol' columns
    data['Stock'] = stock_name
    data['Symbol'] = symbol

    # Append data to the DataFrame
    stock_data = pd.concat([stock_data, data[['Stock', 'Symbol', 'High', 'Low', 'Open', 'Close']]])

# Save the DataFrame to a CSV file
stock_data.to_csv('stock_data.csv', index=True)

# Display the DataFrame
stock_data
```

```python
stockData = pd.read_csv('stock_data.csv')
stockData = stockData.drop([0,1])
stockData.rename(columns={stockData.columns[0]: 'Date'}, inplace=True)
stockData['Date'] = pd.to_datetime(stockData['Date'])

stockData['Date'] = stockData['Date'].dt.date
stockData
```

```python
# careful, convert table like pivot, the following code is from official github combine with ai tools
df_covidconfirm_long = pd.melt(covid19Confirm, id_vars=['Province/State', 'Country/Region','Lat','Long'], var_name='Date', value_name='Confirmed')
df_covidconfirm_long
```

```python
df_coviddeath_long = pd.melt(covid19Death, id_vars=['Province/State', 'Country/Region','Lat','Long'], var_name='Date', value_name='Death')
df_coviddeath_long
```

```python
df_covid_combined = pd.merge(df_covidconfirm_long, df_coviddeath_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'], how='outer')
df_covid_combined
```

```python
# convert stock table
columns = ['Date', 'Stock', 'Symbol', 'Low', 'High', 'Open', 'Close']
stockDataFinal = pd.DataFrame(columns=columns)
base_columns = ['Date', 'Stock', 'Symbol']


additional_columns = stockData.columns[len(base_columns):]

num_columns_per_stock = 4

# loop for each stock basically like str + str, little bit complicate
for i in range(0, len(additional_columns), num_columns_per_stock):
    stock_columns = additional_columns[i : i + num_columns_per_stock]

    if len(stock_columns) == num_columns_per_stock:
        temp_df = stockData[base_columns + list(stock_columns)].copy()
        temp_df.columns = base_columns + ['High', 'Low', 'Open', 'Close']
        stockDataFinal = pd.concat([stockDataFinal, temp_df], ignore_index=True)
stockDataFinal = stockDataFinal.dropna(how='any')
stockDataFinal.isnull().sum()




```

```python
stockDataFinal['Low'] = stockDataFinal['Low'].astype(float)
stockDataFinal['High'] = stockDataFinal['High'].astype(float)
stockDataFinal['Open'] = stockDataFinal['Open'].astype(float)
stockDataFinal['Close'] = stockDataFinal['Close'].astype(float)
stockDataFinal

```

```python

stockDataFinalOrgnized = stockDataFinal.pivot_table(index='Date', columns='Stock', values=['Low', 'High', 'Open', 'Close'])


stockDataFinalOrgnized.columns = [f'{stock}.{timing}' for timing, stock in stockDataFinalOrgnized.columns]


stockDataFinalOrgnized.reset_index(inplace=True)
stockDataFinalOrgnized.to_csv('stockDataFinalOrgnized.csv', index=False)
stockDataFinalOrgnized
```

```python
# pray and hope it works
finalVersionTable = pd.merge(df_covid_combined, stockDataFinalOrgnized, on='Date', how='outer')
finalVersionTable.to_csv('finalVersionTable.csv', index=False)
finalVersionTable
```
