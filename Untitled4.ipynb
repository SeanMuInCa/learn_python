```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Function to draw a flowchart
def draw_flowchart_fixed():
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 12)
    ax.axis('off')

    # Define process boxes
    processes = [
        {"text": "Data Collection\n(Job Data & Resumes)", "xy": (3, 10)},
        {"text": "Data Preprocessing\n(Cleaning & Transformation)", "xy": (3, 8)},
        {"text": "Embedding Generation\n(Sentence-BERT)", "xy": (3, 6)},
        {"text": "Matching Logic\n(Cosine Similarity)", "xy": (3, 4)},
        {"text": "Output & Feedback\n(Job Matches & Suggestions)", "xy": (3, 2)},
    ]

    # Draw rectangles for each process
    for process in processes:
        ax.add_patch(patches.Rectangle(process["xy"], 4, 1, edgecolor='black', facecolor='lightblue'))
        ax.text(process["xy"][0] + 2, process["xy"][1] + 0.5, process["text"], ha='center', va='center', fontsize=10)

    # Draw arrows connecting processes
    arrow_props = dict(facecolor='black', arrowstyle='->')
    for i in range(len(processes) - 1):
        start = (processes[i]["xy"][0] + 2, processes[i]["xy"][1])
        end = (processes[i + 1]["xy"][0] + 2, processes[i + 1]["xy"][1] + 1)
        ax.annotate('', xy=end, xytext=start, arrowprops=arrow_props)

    plt.show()

# Call the function to draw the flowchart
draw_flowchart_fixed()
```
