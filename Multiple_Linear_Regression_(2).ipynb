# **Multiple Linear Regression**

```python
# Step 1 : import library
import pandas as pd
```

dataset link:https://www.kaggle.com/datasets/willianleite/boston-housing-dataset

```python
# Step 2 : import data
house = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Boston.csv')
```

```python
# display first 5 rows
house.head()
```

```python
# display info
house.info()
```

```python
# display summary statistics
house.describe()
```

```python
# Step 3 : define target (y) and features (X)
```

```python
house.columns
```

```python
y = house['MEDV']
```

```python
X = house.drop(['MEDV'],axis=1)   #x1-x13
```

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)

#len(x_train)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_  # m1-m13
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)    #y= c+m1.x1+m2x2+...+m13x13
```

```python
y_pred
```

```python
# Step 8 : model accuracy, further ref: https://www.studytonight.com/post/what-is-mean-squared-error-mean-absolute-error-root-mean-squared-error-and-r-squared
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
